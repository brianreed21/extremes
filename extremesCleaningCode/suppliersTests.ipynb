## Identify Largest Suppliers 
From here, the rate of firms who have supplier weather information should be HIGH! So, we can choose the main suppliers at this point. 

```python
customerYearlyTotals = customers[['year','gvkey','supplier_gvkey','salecs']].drop_duplicates()
print('unique sales data: ', customerYearlyTotals.shape)
customerYearlyTotals['suppliers'] = 1
customerYearlyTotals = customerYearlyTotals.drop(columns = {'supplier_gvkey'}).groupby(['year','gvkey']).sum().reset_index()
print(customerYearlyTotals.shape)
customerYearlyTotals.rename(columns = {'salecs': 'totalSales'}, inplace = True)
customerYearlyTotals.head()
```

```python
customerYearly = customers[['year','gvkey','supplier_gvkey','salecs']].drop_duplicates().merge(customerYearlyTotals)

print(customerYearly.shape)

customerYearly.head()
```

```python
print('We expected to have this many largest suppliers: ', customerYearly[['year','gvkey']].drop_duplicates().shape[0])
```

```python
'''allSuppliersWithWeather.dist.shape
allSuppliersWithWeather_more500k = allSuppliersWithWeather[allSuppliersWithWeather.dist >= 500]
'''
# https://stackoverflow.com/questions/15705630/get-the-rows-which-have-the-max-value-in-groups-using-groupby

idx = customerYearly.groupby(['year','gvkey']).salecs.\
    transform(max) == customerYearly.salecs
largestCustomers = customerYearly[idx].reset_index(drop = True)
# print(largestCustomers.shape)

# find companies who only have one other supplier
singleCustomers = customerYearly[customerYearly.suppliers == 1].reset_index(drop = True)
# print(singleCustomers.shape)

# find largest suppliers of different companies
largestCustomers = largestCustomers.append(singleCustomers).drop_duplicates()
print('And we have this many: ', largestCustomers.shape[0])
```

## Put it Together
Merge the quarterly date with the supplier data (should be ~ a factor of 4, with a little bit of loss), and then with the weather data.

```python
largestSuppliers = largestCustomers.merge(customers)
largestSuppliers.shape
```

Then finally, merge in the weather information.

```python
supplierWeather = customers[['year','qtr','supplier_zipcode']].drop_duplicates().\
    rename(columns = {'supplier_zipcode': 'zipcode'}).astype({'year':       'category',
                           'qtr':        'category',
                           'zipcode':    'category'})
supplierWeather.head()
```

```python
supplierWeather.shape
```

```python
streaks_withLags = pd.read_csv("../../data/companyData/streaks_withLags.csv").\
    drop(columns = {'Unnamed: 0', 'yearQtr'}).astype({'year':       'category',
                           'qtr':        'category',
                           'zipcode':    'category'})


g = pd.read_csv("../../data/companyData/weatherByEstablishment.csv").\
    drop(columns = {"Unnamed: 0"})


allWeather_withLags = pd.read_csv("../../data/companyData/allWeather_withLags.csv").\
    drop(columns = {"Unnamed: 0", 'yearQtr'}).astype({'year':       'category',
                           'qtr':        'category',
                           'zipcode':    'category'})


averages = pd.read_csv("../../data/companyData/quarterlyStatsByZip.csv").\
    drop(columns = {"Unnamed: 0"}).rename(columns = {'ZIP': 'zipcode'})
averages['qtr'] = averages.quarter.str.slice(1,2).astype('float')
averages.drop(columns = {'quarter'}, inplace = True) 
averages = averages.astype({'qtr':        'category',
                           'zipcode':    'category'})


allWeather_withLags2 = pd.read_csv("../../data/companyData/allWeather_withLags_new.csv").\
    drop(columns = {"Unnamed: 0", 'yearQtr'}).astype({'year':       'category',
                           'qtr':        'category',
                           'zipcode':    'category'})


thunderstorms_withLags = pd.read_csv("../../data/companyData/thunderstorms_withLags.csv").\
    drop(columns = {"Unnamed: 0", 'yearQtr'}).astype({'year':       'category',
                           'qtr':        'category',
                           'zipcode':    'category'})
```

```python
supplierWeather = supplierWeather.merge(allWeather_withLags).merge(allWeather_withLags2).\
    merge(averages).merge(streaks_withLags).\
    merge(thunderstorms_withLags)

supplierWeather.head()
```

```python
supplierWeather.rename(columns = {'zipcode': 'supplier_zipcode'}, inplace = True)
```

```python
supplierData = largestSuppliers.merge(supplierWeather)
```
